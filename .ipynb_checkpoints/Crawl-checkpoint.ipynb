```python
import requests
import asyncio
from bs4 import BeautifulSoup
from playwright.async_api import async_playwright
import pandas as pd
import time
import re
```

```python
# web = requests.get(url)
url = "https://dichvucong.gov.vn/p/home/dvc-cau-hoi-pho-bien.html?#mainTitle"
async with async_playwright() as p:
  browser = await p.chromium.launch(headless=True)
  context = await browser.new_context()
  page = await browser.new_page()
  await page.goto(url)
  await page.select_option("select[name='strLeftResult']", "50")
  links_lists=[]
  # Lấy HTML
  while True:
    html = await page.content()
    next_button = await page.query_selector("li.next")
    links_list = BeautifulSoup(html, 'html.parser').find(class_='section-hotro').find(class_='tab-content').find(id='tatCa').find_all('a', class_='item')
    links_lists.append(links_list)
    if next_button and not 'next disabled' in await next_button.get_attribute('class'):
        await next_button.click()
        #await page.wait_for_timeout(2000)
    else:
         break

  await browser.close()

# Kiểm tra nếu yêu cầu thành công
#if response.status_code == 200:
    #print("Yêu cầu thành công!")
#else:
    #print("Lỗi:", response.status_code)
```

```python
data = []
for links_list in links_lists:
  for link in links_list:
    href = link.get('href')
    text = link.get_text(strip=True)
    data.append({'Text':text, 'Link': 'https://dichvucong.gov.vn'+href})
df = pd.DataFrame(data)
```

```python
df = df[df['Text'] != 'undefined']
```

```python
answers = []
TTHC_enformemt = []
for i in range(9000,df.shape[0]):
  #if i==200:
    #await browser.close()
    #break
  url = df.Link[i]
  print("{}: {}".format(i, df.Text[i]))
  async with async_playwright() as p:
    browser = await p.chromium.launch(headless=True)
    context = await browser.new_context()
    page = await browser.new_page()
    await page.goto(url)
    html = await page.content()
    soup = BeautifulSoup(html, 'html.parser')

    try:
      #comment = soup.find(string=re.compile(r'set\(\$answers=\['))
      #content_match = re.search(r'CONTENT=<p>(.*?)</p>', comment)
      #if not content_match:
          #content_match = re.search(r'CONTENT=(.*?)(?:, CONTENT_LOWER|})', comment)
      #answer = content_match.group(1).strip() if content_match else "Không có câu trả lời"
      answer = soup.find(class_="article").find_all('p')[1].get_text(strip=True)
      # Kiểm tra nếu answer là chuỗi rỗng, lấy <p> thứ 3
      if answer == '':
        answer = soup.find(class_="article").find_all('p')[2].get_text(strip=True)
      print(answer)
    except:
      answer = "Không có câu trả lời"
    answers.append(answer)

    TTHC_list_document = {'Tên':[], 'Cơ quan thực hiện':[]}
    try:
      TTHC_links_list = soup.find('ul',class_='list-document').find_all('li')

      for TTHC_link in TTHC_links_list:
        TTHC_link = "https://dichvucong.gov.vn/p/home/" + TTHC_link.find('a').get('href')
        await page.goto(TTHC_link)
        TTHC_html = await page.content()
        TTHC_soup = BeautifulSoup(TTHC_html, 'html.parser')
        TTHC_document = {'Tên':None, 'Cơ quan thực hiện':[]}

        try:
          tthc_name = TTHC_soup.find('h1', class_='main-title -none').get_text(strip=True)
          tthc_agency_section = TTHC_soup.find('h2', string="Cơ quan thực hiện")

          if tthc_agency_section:
              for sibling in tthc_agency_section.find_next_siblings():
                  if sibling.name != 'div' or 'article' not in sibling.get('class', []):
                      break
                  TTHC_document['Cơ quan thực hiện'].append(sibling.get_text(strip=True).replace('\xa0', ' '))
          else:
              print("Không tìm thấy tiêu đề 'Cơ quan thực hiện'")
          # Lưu thông tin vào dictionary
          TTHC_document['Tên'] = tthc_name
        except:
          print(f"Không thể lấy thông tin từ: {TTHC_link}")

        TTHC_list_document['Tên'].append(TTHC_document['Tên'])
        TTHC_list_document['Cơ quan thực hiện'].append(TTHC_document['Cơ quan thực hiện'])

    except:
      print(f"Không thể lấy link thông tin TTHC từ: {url}")
    TTHC_enformemt.append(TTHC_list_document)
    await browser.close()
```

```python
TTHC_save = TTHC_enformemt
answers_save = answers
```

```python
df_save_4 = df.iloc[9000:]# tu cau 5760 toi 6999
df_save_4.shape
```

```python
df_save_4.loc[:, 'Câu trả lời'] = answers_save
df_save_4.loc[:, 'TTHC'] = TTHC_save
```
