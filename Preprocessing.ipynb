convert TTHC to dict

```python
df['TTHC'] = df['TTHC'].apply(ast.literal_eval)
```

Divide TTHC into Tên_TTHC and Tên_cơ_quan_thực_hiện

```python
df['Tên_TTHC'] = df['TTHC'].apply(lambda x: x['Tên'] if isinstance(x, dict) and 'Tên' in x else None)
df['Tên_cơ_quan_thực_hiện'] = df['TTHC'].apply(lambda x: x['Cơ quan thực hiện'] if isinstance(x, dict) and 'Cơ quan thực hiện' in x else None)
```

```python
df_nanNameTTHC = df[df['Tên_TTHC'].apply(lambda x: x == [])]
df_nanAgencyTTHC = df[df.Tên_cơ_quan_thực_hiện.apply(lambda x: x == [])]
df_nanAgencyTTHC.shape
if df_nanAgencyTTHC.equals(df_nanNameTTHC):
    print('true')
else:
    print('false')
```
