```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Preprocessed Data.csv')
```

```python
data['TTHC'] = data['TTHC'].apply(ast.literal_eval)
```

```python
TTHC = []
for i in range(data.shape[0]):
    for item in data.TTHC[i]['Tên']:
        if item not in TTHC:
            print(item)
            TTHC.append(item)


```

```python
len(TTHC)
```

```python
data_TTHC = data[data['TTHC'].apply(lambda x: any(item in TTHC for item in x.get('Tên', [])))]
```

```python
data_TTHC.reset_index(inplace=True, drop=True)
data_TTHC
```

```python
TTHC_Name = []
TTHC_Link = []

async def fetch_data():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context()
        page = await context.new_page()

        max_retries = 3  # Số lần thử lại tối đa

        for i in range(data_TTHC.shape[0]):
            url = data_TTHC.Link[i]
            retries = 0

            while retries < max_retries:
                try:
                    print(f"Đang crawl hàng {i} - Thử lần {retries + 1}: {url}")
                    await page.goto(url, timeout=10000)
                    html = await page.content()
                    soup = BeautifulSoup(html, 'html.parser')

                    tthc_list = soup.find('ul', class_='list-document')
                    if tthc_list is None:
                        raise ValueError(f"Không tìm thấy 'list-document' trong URL {url}")

                    links = tthc_list.find_all('a')
                    result = [(link.text.strip(), link['href']) for link in links]

                    for name, link in result:
                        print(name, link)
                        TTHC_Name.append(name)
                        TTHC_Link.append(link)

                    break

                except Exception as e:
                    print(f" Lỗi tại hàng {i} (Lần {retries + 1}): {e}")
                    retries += 1
                    await asyncio.sleep(2)

                    if retries == max_retries:
                        print(f"Không thể crawl dữ liệu tại hàng {i} sau {max_retries} lần thử")
                        TTHC_Name.append(None)
                        TTHC_Link.append(None)

        await browser.close()

await fetch_data()
```

```python
data_TTHC_df = pd.DataFrame({'Tên': TTHC_Name, 'Link': TTHC_Link})
data_TTHC_df
```

```python
data_TTHC_new = data_TTHC_df.drop_duplicates(keep='first')
data_TTHC_new.reset_index(inplace=True,drop = True)
data_TTHC_new
```

```python
#them utf
pd.DataFrame(data_TTHC_new).to_csv('TTHC.csv', index=False, encoding='utf-8-sig')
```

### Crawl Data TTHC

```python
TTHC_df = pd.read_csv('TTHC.csv')
TTHC_df.head()
```

```python
Chitiet_TTHC = []
async def fetch_TTHC_data():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context()
        page = await context.new_page()

        max_retries = 3  # Số lần thử lại tối đa

        for i in range(TTHC_df.shape[0]):
            url = "https://dichvucong.gov.vn/p/home/" + TTHC_df.Link[i]
            retries = 0

            while retries < max_retries:
                try:
                    print(f"Đang crawl hàng {i} - Thử lần {retries + 1}: {url}")
                    await page.goto(url, timeout=10000)

                    await page.wait_for_selector(".col-sm-8.col-xs-12", timeout=5000)
                    text_content = await page.inner_text(".col-sm-8.col-xs-12")
                    print(f"Nội dung: {text_content}")
                    Chitiet_TTHC.append(text_content)
                    break




                except Exception as e:
                    print(f"Lỗi tại hàng {i} (Lần {retries + 1}): {e}")
                    retries += 1
                    await asyncio.sleep(2)

                    if retries == max_retries:
                        print(f"Không thể crawl dữ liệu tại hàng {i} sau {max_retries} lần thử")
                        Chitiet_TTHC.append(None)

        await browser.close()

await fetch_TTHC_data()
```

```python
TTHC_df = pd.DataFrame({'Tên': TTHC_df.Tên, 'Link': TTHC_df.Link, 'ChiTiet': Chitiet_TTHC})
TTHC_df
```

```python
pd.DataFrame(TTHC_df).to_csv('ChiTietTTHC.csv', index=False, encoding='utf-8-sig')
```
